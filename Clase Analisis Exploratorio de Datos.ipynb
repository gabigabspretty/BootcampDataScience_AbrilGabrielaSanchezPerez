```python
import requests
base_url = "https://api-colombia.com/"
endpoint = "api/v1/Department"
response = requests.get(base_url + endpoint)

if response.status_code == 200:
  print("Conexión exitosa a la API. Codigo de estado:", response.status_code)
else:
  print("Conexión fallida a la API. Codigo de estado", response.status_code)
```

```python
data = response.json()
print('Top 10 Departamentos de Colombia:')
for item in data [:10]:
    print(item['name'], ":", item['description'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('tips')

df.head()
```

```python
sns.barplot(data = df, x= "time", y="tip")
```

```python
sumatoria_propinas = df.groupby('time')['tip'].mean()
plt.bar(sumatoria_propinas.index, sumatoria_propinas.values)
plt.title('Promedio de propinas por hora del dia')
plt.xlabel('Hora del dia')
plt.ylabel('Promedio de propinas')
```
