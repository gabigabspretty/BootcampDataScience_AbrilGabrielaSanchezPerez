# Parte 1.  Python (Extracción y preparación)
### API seleccionada: https://pokeapi.co/

La idea propuesta para estas consultas es que me permita saber las características de cada Pokemon, así como en el primer taller donde hice un análisis de Pokemon para poder usarlo en el juego TCG, ahora con todas sus caracteristicas.

```python
# Paso 1. Importar las librerias a usar, así como consultar la API seleccionada.
import pandas as pd
import requests
```

```python
# Paso 2. Consultar la PokeAPI y crear un DataFrame para hacer las consultas. 

# Asignación de 'bolsa' para almacenar todos los datos a consultar.
mi_pokedex = []

# Ahora hacer la asignacion de los datos en la url de la API, datos a consultar = 150 pokemones.
for i in range(1, 151):
    url = f'https://pokeapi.co/api/v2/pokemon/{i}'
    response = requests.get(url)
    data = response.json()

# Crear la segmentación de datos que necesito para hacer las visualizaciones.
    pokemon_data = {
        'id': data['id'],
        'nombre': data['name'],
        'altura': data['height'],
        'peso': data['weight'],
        'tipo_1': data["types"][0]["type"]["name"],
        'tipo_2': data["types"][1]["type"]["name"] if len(data["types"]) > 1 else "ninguno",
        'hp': data["stats"][0]["base_stat"],
        'ataque': data["stats"][1]["base_stat"],
        'defensa': data["stats"][2]["base_stat"],
        'velocidad': data["stats"][5]["base_stat"],
    }
    mi_pokedex.append(pokemon_data) # Esto sirve para poder agregar la informacion en la Pokedex que acabo de crear
print(f"Estos son los Pokemon que he atrapado: {len(mi_pokedex)}") # Para poder verificar que traje la info correcta
```

```python
# Paso 3. Ya tengo los datos, pero ahora necesito convertirlos en Datafram (df) para consultar todo

df_pokedex = pd.DataFrame(mi_pokedex) # Usando la libreria de pandas (pd) obtengo una tabla al estilo Excel.

# Ahora para verificar que esta tabla si haga sentido, usaré la funcion head()

df_pokedex.head()
```

```python
# Paso 4. (Y ultimo dentro de Python) Creacion de el archivo .csv

# Usando la funcion .to_csv() en panda, podré exportar el archivo y llevarlo a Power BI. 

df_pokedex.to_csv('pokedex.csv', index=False) # index=False me evita agregar esa primera columna que no sirve (solo es informativa de la posicion del dato)

print ("Archivo .csv generado correctamente, nos vemos en Power BI")
```
