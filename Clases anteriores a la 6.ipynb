```python
### Que se ingrese una clave y usuario correcto
### Si tiene el dinero suficiente para retirar, sino que muestre un mensaje de error

saldo = 45000000

usuario_base = "abril2909"
contraseña_base = "Abril202603"

usuario = input("Ingrese su usuario")
contraseña = input("Ingrese su contraseña")
retiro = float(input("Monto a retirar"))

try:
    if usuario != usuario_base or contraseña!= contraseña_base:
        raise ValueError("Usuario o contraseña incorrecta")
    elif retiro <= 0:
        raise ValueError("El monto a retirar debe ser mayor a cero")
    elif retiro % 1000!= 0:
        raise ValueError ("El monto a retirar debe ser un multiplo de 1000")
    elif retiro < saldo:
        gravamen = retiro * 4/10000
        if retiro +gravamen > saldo:
            raise ValueError("Saldo Insuficiente para cubrir retiro y gravamen financiero")
        else:
            #saldo -= (retiro + gravamen)
            saldo = saldo - (retiro + gravamen) 
            print(f"Retiro exitoso- Saldo restante: ${saldo}")

except Exception as e:
    print(e)




```

ACTIVIDAD
Validador de contraseñas

Descripción:

Solicite una contraseña al usuario
Valide según múltiples condiciones
Maneje errores/excepciones
Requerimientos:

Pedir la contraseña por input
Validar que tenga al menos 8 caracteres. Si no se cumple, imprimir "Error: Minimo 8 caracteres"
Contener al menos un número. Si no se cumple, imprimir "Error: Falta numero"
Incluir al menos una mayúscula. Si no se cumple, imprimir "Error: Falta mayuscula"
Si todas las condiciones se cumplen, imprimir "Clave valida"
Implementar:

Estructuras if-elif-else
Manejo de excepciones con try-except
Uso de raise para errores personalizados

```python
""""
Generador de Contraseñas:

Recuerde que debe contener como mínimo:
- 8 Caracteres
- 1 Mayuscula
- 1 número

"""
try:
    password = input ("Ingrese su contraseña: ")

    #Valide que existan mínimo 8 caracteres
    if len(password) < 8:
        raise ValueError("Recuerde que debe contener minimo 8 Caracteres")
    #Valide que contenga al menos 1 numero
    tiene_numero = any(char.isdigit() for char in password)
    if not tiene_numero:
        raise ValueError("Error: Debe contener al menos 1 numero")
    #Valide que contenga al menos 1 mayúscula
    tiene_mayuscula=any(char.isupper()for char in password)
    if not tiene_mayuscula:
        raise ValueError("Error: Debe contener al menos 1 mayuscula")

    #Si todo está bien
    print("Clave valida")
except ValueError as e:
    print(e)
```
